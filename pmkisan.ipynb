```python
r1.columns = (
    r1.columns
      .str.strip()
      .str.lower()
      .str.replace('\n', ' ', regex=False)
)
r1.columns
```

To ensure robustness, we identify relevant variables by their position in the dataset after inspection.

1. state_clean stores standardized state names

2. beneficiaries captures the number of PM-KISAN beneficiaries

3. amount_released captures total PM-KISAN disbursement (â‚¹ crore)

Numeric fields are converted from text to numeric format after removing commas.

```python
r1['state/ut'] = r1.iloc[:, 1].astype(str).str.strip().str.lower()

r1['beneficiaries'] = (
    r1.iloc[:, 2]
    .astype(str)
    .str.replace(',', '', regex=False)
    .astype(float)
)

r1['amount_released'] = (
    r1.iloc[:, 3]
    .astype(str)
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

Verifying cleaned variables

```python
r1[['state/ut', 'beneficiaries', 'amount_released']].head()
```

Aggregating beneficiary counts and disbursement amounts across all rows belonging to the same state

```python
state_r1 = (
    r1.groupby('state/ut')
      .agg(
          total_beneficiaries=('beneficiaries', 'sum'),
          total_amount=('amount_released', 'sum')
      )
      .reset_index()
)
## final state metric
state_r1['amount_per_beneficiary'] = ( state_r1['total_amount'] / state_r1['total_beneficiaries'])
```
